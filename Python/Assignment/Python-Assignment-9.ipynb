#### Question 1: Plot the min and max temperature for a year

```python
import matplotlib.pyplot as pyt
import pandas as pd
import numpy as np
from scipy import optimize
```

```python
temp_max = np.array([39,41,43,47,49,51,45,38,37,29,27,25])
temp_min = np.array([21,23,27,28,32,35,31,28,21,19,17,18])
months = np.array(range(0,12))

def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      temp_max, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      temp_min, [-40, 20, 0])

days = np.linspace(0, 12, num=365)

pyt.figure()
pyt.plot(months, temp_max, 'ro')
pyt.plot(days, yearly_temps(days, *res_max), 'r-')
pyt.plot(months, temp_min, 'bo')
pyt.plot(days, yearly_temps(days, *res_min), 'b-')
pyt.xlabel('Month')
pyt.ylabel('Temperature (Â°C)')

pyt.show()
```

#### Question 2: Plot the titanic data using the matplotlib

```python
url = "https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)

count_female = len(titanic[titanic['sex']=='female'])
count_male = len(titanic[titanic['sex']=='male'])
pyt.pie((count_female,count_male),labels = ('female','male'), colors = ('pink','blue'))
```

```python
pyt.rcParams['figure.figsize'] = (20,10)
```
