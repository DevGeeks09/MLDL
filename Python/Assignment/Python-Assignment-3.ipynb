```python
"""1.1 Implement an own version of reduce method"""

def myreduce(func,lst):
    if type(lst) == set:
        lst = list(lst)
    elif type(lst) == dict:
        lst = list(lst.keys())
        
    if len(lst) == 1:
        return lst[0]
    
    result  = func(lst[0], lst[1])
    for x in range(2,len(lst)):
        result = func(result, lst[x])
        
    return result        
```

```python
"""1.2 Implement an own version of filter method"""

def myfilter(func,lst):
    result = list()
    for x in lst:
        ret = func(x)
        if ret == False or ret==None:
            continue
            
        result.append(x)
            
    return result       
```

```python
"""2. Implement List comprehensions"""

"""['A', 'C', 'A', 'D', 'G', 'I', ’L’, ‘ D’] """
lst = "acadgild"
print("Scenario 1: ", [x.upper() for x in lst])

"""['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']"""
def cubeAlph(num):
    for y in ['x','y','z']:
        for i in range(1,num+1):
            yield y*i

print("Scenario 2: ", [n for n in cubeAlph(4)])

"""['x', 'y', 'z', 'xx', 'yy', 'zz', 'xxx', 'yyy', 'zzz', 'xxxx', 'yyyy', 'zzzz']"""
def cubeAlph2(num):
    for y in range(1,num+1):
        for i in ['x','y','z']:
            yield i*y

print("Scenario 3: ", [n for n in cubeAlph2(4)])

"""[[2], [3], [4], [3], [4], [5], [4], [5], [6]]"""
"""[[2, 3, 4, 5], [3, 4, 5, 6], [4, 5, 6, 7], [5, 6, 7, 8]]"""

def printList(num):
    value = 2
    for i in range(num):
        for j in range(3):
            yield [value+j]
        value = value+1
            
print("Scenario 4: ",[x for x in printList(3)])

def printList2(num):
    value = 2
    for i in range(num):
        output = list()
        for j in range(4):
            output.append(value+j)
        yield output
        value = value+1
            
print("Scenario 5: ",[x for x in printList2(4)])

""" [(1, 1), (2, 1), (3, 1), (1, 2), (2, 2), (3, 2), (1, 3), (2, 3), (3, 3)]"""
output = list()
for i in range(3):
    for j in range(3):        
        output.append(tuple([j+1,i+1]))
        
print("Scenario 6: ",output)
```
