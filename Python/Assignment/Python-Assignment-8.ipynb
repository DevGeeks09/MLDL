#### Question: Plot 3 dimensions 3d chart from iris data

```python
import matplotlib.pyplot as plt
from sklearn import datasets

%matplotlib notebook

irisdf = datasets.load_iris()
```

```python
fig = plt.figure()
axe = plt.axes(projection='3d')
xaxis = irisdf.data[:,0]
yaxis = irisdf.data[:,1]
zaxis = irisdf.data[:,2]
axe.scatter3D(xaxis,yaxis,zaxis, s=30, c=irisdf.target)

axe.set_title("IRIS Dataset Scatter Plot")
axe.set_xlabel(irisdf.feature_names[0])
axe.set_ylabel(irisdf.feature_names[1])
axe.set_zlabel(irisdf.feature_names[2])
```
