### Fetch the data from the sql using pandas

```python
with open('sample-csv/addresses.csv') as f:
    data = csv.reader(f, delimiter = '\n')
    for lines in enumerate(data):
        if data.line_num == 1:
            continue
        for i in (lines[1]):
            list_ = i.split(',')
            cursor.execute(insert_add_tb_query.format(value1 = (list_[0]), value2 = (list_[1]), value3 = (list_[2])
                            , value4 = (list_[3]), value5 = (list_[4]), value6 = (list_[5]), value7 = (list_[6])
                            , value8 = (list_[7])))
    mydb.commit()
```
